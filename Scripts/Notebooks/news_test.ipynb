```python
from Modules.params import get_params
from top2vec import Top2Vec
from os.path import join
```

```python
params=get_params()
filename=join(params["path models"],
              params["news model"])
model = Top2Vec.load(filename)
```

```python
topic_words, word_scores, topic_scores, topic_nums = model.search_topics(keywords=["president"], 
                                                                         num_topics=5)
for topic in topic_nums:
    model.generate_topic_wordcloud(topic)
```

```python
model.topic_words[48]
```

```python
documents, document_scores, document_ids = model.search_documents_by_topic(topic_num=48, 
                                                                           num_docs=5)
for doc, score, doc_id in zip(documents, document_scores, document_ids):
    print(f"Document: {doc_id}, Score: {score}")
    print("-----------")
    print(doc)
    print("-----------")
    print()
```

```python
documents, document_scores, document_ids = model.search_documents_by_keywords(keywords=["president", "privacy"], num_docs=5)
for doc, score, doc_id in zip(documents, document_scores, document_ids):
    print(f"Document: {doc_id}, Score: {score}")
    print("-----------")
    print(doc)
    print("-----------")
    print()
```

```python
words, word_scores = model.similar_words(keywords=["bush"], keywords_neg=[], num_words=20)
for word, score in zip(words, word_scores):
    print(f"{word} {score}")
```
